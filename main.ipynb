```python
import requests
import json

# Replace with your actual API key
API_KEY =  'q1XdYSQwrujtCGDkQztRkd2cVNuxutzxk1SK81HhI60'# Insert your actual Rebase Energy API key

# API endpoint
url = "https://api.rebase.energy/weather/v2/query"

# Headers with the correct Content-Type
headers = {
    "Authorization": API_KEY,
    "Content-Type": "application/json"
}

# Parameters for the weather forecast
params = {
    "model": "DWD_ICON-EU",  # Example model
    "start-date": "2025-06-02",
    "end-date": "2025-06-05",
    "latitude": "59.3293",   # Stockholm
    "longitude": "18.0686",
    "variables": "Temperature,WindSpeed, SolarDownwardRadiation",
    "output-format": "json",
    'output-schema': 'list'
}

# Send POST request with JSON payload
response = requests.post(url, headers=headers, data=json.dumps(params))

# Check and print the response
if response.ok:
    print(response.json())
else:
    print(f"Error {response.status_code}: {response.text}")
```
